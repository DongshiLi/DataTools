```python
class OHLC:
    
    def __init__(self, df, PeriodCount):
        
        self.df = df
        self.PeriodCount = PeriodCount
        pass
    
    def GetOpenPrice(self):
        
        OpenDict = dict()
        
        OpenTime1 = datetime(datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').year,
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').month, 
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').day, 9, 15, 0)
        OpenTime2 = datetime(datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').year,
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').month, 
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').day,13, 0, 0)
        
        for i in range(int(8100 / self.PeriodCount)):
            for j in range(len(self.df['TimeStamp'])):
                Time = datetime.strptime("%s %s" % (str(self.df['TradingDay'][j]), self.df['TimeStamp'][j]), '%Y%m%d %H%M%S')
                #print(Time, OpenTime1)
                if((Time > OpenTime1 + timedelta(seconds=i * self.PeriodCount)) & 
                  (Time <= OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))):
                    if(len(OpenDict) < i + 1):
                        if(self.df['LastPrice'][j] == 0):
                            OpenDict[str(OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))] = (self.df['BidPrice1'][j] + self.df['AskPrice1'][j]) / 2
                        else:
                            OpenDict[str(OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))] = self.df['LastPrice'][j]
        
        for i in range(int(7200 / self.PeriodCount)):
            for j in range(len(self.df['TimeStamp'])):
                Time = datetime.strptime("%s %s" % (str(self.df['TradingDay'][j]), self.df['TimeStamp'][j]), '%Y%m%d %H%M%S')
                if((Time > OpenTime2 + timedelta(seconds=i * self.PeriodCount)) & 
                  (Time <= OpenTime2 + timedelta(seconds=(i + 1) * self.PeriodCount))):
                    if(len(OpenDict) < i + 1):
                        if(self.df['LastPrice'][j] == 0):
                            OpenDict[str(OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))] = (self.df['BidPrice1'][j] + self.df['AskPrice1'][j]) / 2
                        else:
                            OpenDict[str(OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))] = self.df['LastPrice'][j]
        
        return OpenDict
    
    def GetHighPrice(self):
        
        HighDict = dict()
        
        OpenTime1 = datetime(datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').year,
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').month, 
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').day, 9, 15, 0)
        OpenTime2 = datetime(datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').year,
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').month, 
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').day,13, 0, 0)
        
        for i in range(int(8100 / self.PeriodCount)):
            HighPrice = 0
            Price = 0
            for j in range(len(self.df['TimeStamp'])):
                Time = datetime.strptime("%s %s" % (str(self.df['TradingDay'][j]), self.df['TimeStamp'][j]), '%Y%m%d %H%M%S')
                if((Time > OpenTime1 + timedelta(seconds=i * self.PeriodCount)) & 
                  (Time <= OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))):
                    if(self.df['LastPrice'][j] == 0):
                        Price = (self.df['BidPrice1'][j] + self.df['AskPrice1'][j]) / 2
                        if(HighPrice < Price):
                            HighPrice = Price
                    else:
                        Price = self.df['LastPrice'][j]
                        if(HighPrice < Price):
                            HighPrice = Price
            if(HighPrice == 0) :
                pass
            else:
                HighDict[str(OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))] = HighPrice
            
        for i in range(int(7200 / self.PeriodCount)):
            HighPrice = 0
            Price = 0
            for j in range(len(self.df['TimeStamp'])):
                Time = datetime.strptime("%s %s" % (str(self.df['TradingDay'][j]), self.df['TimeStamp'][j]), '%Y%m%d %H%M%S')
                if((Time > OpenTime2 + timedelta(seconds=i * self.PeriodCount)) & 
                  (Time <= OpenTime2 + timedelta(seconds=(i + 1) * self.PeriodCount))):
                    if(self.df['LastPrice'][j] == 0):
                        Price = (self.df['BidPrice1'][j] + self.df['AskPrice1'][j]) / 2
                        if(HighPrice < Price):
                            HighPrice = Price
                    else:
                        Price = self.df['LastPrice'][j]
                        if(HighPrice < Price):
                            HighPrice = Price
            if(HighPrice == 0) :
                pass
            else:
                HighDict[str(OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))] = HighPrice
                
        return HighDict
    
    def GetLowPrice(self):
        
        LowDict = dict()
        
        OpenTime1 = datetime(datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').year,
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').month, 
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').day, 9, 15, 0)
        OpenTime2 = datetime(datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').year,
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').month, 
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').day,13, 0, 0)
        
        for i in range(int(8100 / self.PeriodCount)):
            LowPrice = 9999
            Price = 0
            for j in range(len(self.df['TimeStamp'])):
                Time = datetime.strptime("%s %s" % (str(self.df['TradingDay'][j]), self.df['TimeStamp'][j]), '%Y%m%d %H%M%S')
                if((Time > OpenTime1 + timedelta(seconds=i * self.PeriodCount)) & 
                  (Time <= OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))):
                    if(self.df['LastPrice'][j] == 0):
                        Price = (self.df['BidPrice1'][j] + self.df['AskPrice1'][j]) / 2
                        if(LowPrice > Price):
                            LowPrice = Price
                    else:
                        Price = self.df['LastPrice'][j]
                        if(LowPrice > Price):
                            LowPrice = Price
            if(LowPrice == 9999) :
                pass
            else:
                LowDict[str(OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))] = LowPrice
        
        for i in range(int(7200 / self.PeriodCount)):
            LowPrice = 9999
            Price = 0
            for j in range(len(self.df['TimeStamp'])):
                Time = datetime.strptime("%s %s" % (str(self.df['TradingDay'][j]), self.df['TimeStamp'][j]), '%Y%m%d %H%M%S')
                if((Time > OpenTime2 + timedelta(seconds=i * self.PeriodCount)) & 
                  (Time <= OpenTime2 + timedelta(seconds=(i + 1) * self.PeriodCount))):
                    if(self.df['LastPrice'][j] == 0):
                        Price = (self.df['BidPrice1'][j] + self.df['AskPrice1'][j]) / 2
                        if(LowPrice > Price):
                            LowPrice = Price
                    else:
                        Price = self.df['LastPrice'][j]
                        if(LowPrice > Price):
                            LowPrice = Price
            if(LowPrice == 9999) :
                pass
            else:
                LowDict[str(OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))] = LowPrice
        
        return LowDict
    
    def GetClosePrice(self):
        
        CloseDict = dict()
        
        OpenTime1 = datetime(datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').year,
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').month, 
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').day, 9, 15, 0)
        OpenTime2 = datetime(datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').year,
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').month, 
                             datetime.strptime(str(self.df['TradingDay'][0]), '%Y%m%d').day,13, 0, 0)
        
        for i in range(int(8100 / self.PeriodCount)):
            ClosePrice = 0
            for j in range(len(self.df['TimeStamp'])):
                Time = datetime.strptime("%s %s" % (str(self.df['TradingDay'][j]), self.df['TimeStamp'][j]), '%Y%m%d %H%M%S')
                #print(Time, OpenTime1)
                if((Time > OpenTime1 + timedelta(seconds=i * self.PeriodCount)) & 
                  (Time <= OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))):
                    if(self.df['LastPrice'][j] == 0):
                        ClosePrice = (self.df['BidPrice1'][j] + self.df['AskPrice1'][j]) / 2
                    else:
                        ClosePrice = self.df['LastPrice'][j]
            if(ClosePrice == 0):
                pass
            else:
                CloseDict[str(OpenTime1 + timedelta(seconds=(i + 1) * self.PeriodCount))] = ClosePrice
        
        return CloseDict
```

```python
def MergeDict(dict1, dict2):
    for key in dict1.keys():
        a=[]
        b=[]
        b = dict1[key]    
        for i in b:
            a.append(i)
        a.append(dict2[key])  
        dict1[key] = a
    return dict1
```
